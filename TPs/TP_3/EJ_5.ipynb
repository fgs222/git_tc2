###     Teoría de Circuitos II

Ejercicio 5

El siguiente diagrama de Bode corresponde a la respuesta en módulo de la transferencia de una red de énfasis, utilizada en un transmisor de FM para Broadcasting. Diseñar el circuito, verificando el mismo mediante simulación

- $\omega_{1} = 2\pi f_{1} ~\Rightarrow~ \omega_{1} = 4000\pi~s^{-1}$
- $\omega_{2} = 2\pi f_{2} ~\Rightarrow~ \omega_{2} = 64000\pi~s^{-1}$

Analisis de pendiente:

$$
\log(\frac{f_{1}}{f_{2}}) = \log(\frac{64}{4}) = 1,2dec ~\Rightarrow~ 18db-(-6db)=24db  
$$

- $1,2 dec ~\Rightarrow~ 24db~~~~~~~~~~~~~~~~~~~~~~~~~~$Pendiente de 20db/dec, cero simple en $\omega_{1}$ y polo simple en $\omega_{2}$
- $1 dec ~~~~~\Rightarrow~ 20db$

$$
H(s) = K \frac{s+4000\pi}{s+64000\pi}
$$

Analisis de continua:

En $\omega=0$ hay ganancia de $6db ~\Rightarrow~ H(0) = 10^{\frac{-6}{20}} = \frac{1}{2}$

$$
H(0) = K \frac{4000\pi}{64000\pi} = \frac{1}{2} ~\Rightarrow~ K = 8
$$

Verificando en alta frecuencia: 
$$
20\log(8)=18db
$$

Normalizando con $~\Omega_{w}=4000\pi$:

- $\omega_{1}=1$
- $\omega_{2}=16$

$$
H(s) = 8\frac{s+1}{s+16}
$$

Proponiendo el circuito a continuacion:

$$
H(s) = -\frac{y1}{y2} = \frac{C_{1}}{C_{2}} \frac{s+\frac{1}{R_{1}C_{1}}}{s+\frac{1}{R_{2}C_{2}}}
$$

$$
K = \frac{C_{1}}{C_{2}} ~\Rightarrow~ C_{1} = 8C_{2}~;~Propongo~~C_{1} = 8,~C_{2}=1
$$

$$
\frac{1}{R_{1}C_{1}} = 1 ~\Rightarrow~ 1 = \frac{1}{R_{1}8} ~\Rightarrow~ R_{1}= \frac{1}{8}
$$

$$
\frac{1}{R_{2}C_{2}} = 16 ~\Rightarrow~ 16 = \frac{1}{R_{2}} ~\Rightarrow~ R_{2}=\frac{1}{16}
$$

Desnormalizando los componentes:

```python
import math

# Componentes normalizados
C1_n  = 8
R1_n  = 0.125
C2_n = 1
R2_n = 0.0625

# Normas
N_omega = 2 * math.pi * 2000     # Norma de frecuencia
N_z = 16931.37692                # Norma de impedancia para C1=4.7nF

# Desnormalizacion
C1 = C1_n / (N_omega * N_z)
R1 = R1_n * N_z
C2 = C2_n / (N_omega * N_z) 
R2 = R2_n * N_z 

print('C1: {}; R1: {}; C2: {}; R2: {}'.format(C1,R1,C2,R2))
```
