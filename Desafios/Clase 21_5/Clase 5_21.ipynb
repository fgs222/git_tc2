Se debe diseñar un filtro pasaaltos , que presente máxima planicidad en la banda de paso (frecuencia de corte = 300Hz)  y un cero  de transmisión en 100 Hz. El prototipo pasabajos normalizado presenta la siguiente respuesta:

- $F_{c}$ = 300 Hz

- Determine la expresión de H(s) del filtro pasaltos normalizado 

Del grafico es posible determinar que el cero de transmision se encuentra en $\omega=3$. Como la transferencia de un filtro elimina banda tiene 2 ceros complejos conjugados sabemos que el numerador responde a la forma $~\omega^{2}+\omega_{z}^{2}$, por lo que el numerador de la transferencia a determinar es: $~\omega^{2}+3^{2}$

En alta frecuencia la transferencia decae 20 db/dec por lo que es posible determinar que el limite tendiendo al infinito de la transferencia a determinar es:

$$
\lim_{s \to \infty}\frac{1}{s}
$$

En conclusion, el orden del denominador de la transferencia debe ser de orden N=3. Debido a que el filtro deseado es del tipo Butterworth, los polos de la transferencia deberan estar equidistantes a un angulo de $\frac{\pi}{3}$, por lo que el denominador del filtro deseado tendra el siguiente denominador: $(s+1)(s^{2}+s+1)$

Es posible conformar la transferencia deseada de la siguiente forma:

$$
H_{lp}(s)=k\frac{s^{2}+3^{2}}{(s+1)(s^{2}+s+1)}
$$

Para determinar el valor de la constante k, evaluamos la funcion en 0:

$$
H_{lp}(0) = 9k = 1~~~~\Rightarrow~~~~k=\frac{1}{9}
$$

Por lo que la transferencia pedida queda determinada de la siguietne forma:

$$
H_{lp}(s)=\frac{1}{9}\frac{s^{2}+9}{s^{3}+2s^{2}+2s+1}
$$

Aplicando el siguiente kernel de transformacion: $s_{hp} = \frac{1}{s_{lp}}$

Aplicando el kernel a la transformacion obtenemos:

$$
H_{hp}(s)=\frac{1}{9}\frac{9s^{3}+s}{s^{3}+2s^{2}+2s+1}
$$
