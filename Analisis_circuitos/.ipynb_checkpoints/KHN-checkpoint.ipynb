# KHN

Ecuaciones:

$$
\frac{V_{a}}{R_{f1}} = -sCV_{b}~~~~~(1)
$$

$$
\frac{V_{b}-V_{x}}{R_{4}} = \frac{V_{b}}{R_{q}}~~~~~(2)
$$

$$
\frac{V_{b}}{R_{f2}} = -sCV_{c}~~~~~(3)
$$

$$
\frac{V_{in}}{R_{g}} = \frac{V_{x}-V_{c}}{R_{1}} + \frac{V_{x}-V_{a}}{R_{2}}~~~~~(4)
$$

De (3):
$$
V_{b} = -sCR_{f2}V_{C}
$$

De (1):
$$
V_{a} = -sCR_{f1}V_{b}
$$

De (2):
$$
\frac{V_{x}}{R_{4}} = \frac{V_b}{R_{4}} - \frac{V_{b}}{R_{q}}~~~\Rightarrow~~~\frac{V_{x}}{R_{4}} = \frac{(R_{q}-R_{4})}{R_{4}R_{q}}V_{b}~~~\Rightarrow~~~V_{x} = \frac{R_{q}-R_{4}}{R_{q}}V_{b}
$$

## Low Pass Filter

$$
\frac{V_{in}}{R_{g}} = \frac{V_{x}}{R_{1}}-\frac{V_{c}}{R_{1}}+\frac{V_{x}}{R_{2}}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = \frac{R_{1}(R_{q}-R_{4})}{R_{q}}V_{b}-\frac{V_{c}}{R_{1}}+\frac{V_{x}}{R_{2}}+\frac{sCR_{f1}}{R_{2}}V_{b}
$$

$$
\frac{V_{in}}{R_{g}} = (\frac{R_{1}(R_{q}+R_{4})}{R_{q}}-\frac{sCR_{f1}}{R_{2}})V_{b}-\frac{V_{c}}{R_{1}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}}V_{b}
$$

$$
\frac{V_{in}}{R_{g}} = (\frac{R_{1}(R_{q}-R_{4})}{R_{q}}+\frac{sCR_{f1}}{R_{2}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}})V_{b}-\frac{V_{c}}{R_{1}}
$$

$$
\frac{V_{in}}{R_{g}} = -(\frac{R_{1}(R_{q}-R_{4})}{R_{q}}+\frac{sCR_{f1}}{R_{2}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}})sCR_{f2}V_{C}-\frac{V_{c}}{R_{1}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{sCR_{f1}R_{q}+R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{R_{q}R_{2}})sCR_{f2}V_{C}-\frac{V_{c}}{R_{1}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{s^{2}C^{2}R_{f1}R_{f2}R_{q}+sCR_{2}R_{f2}(R_{1}+R_{2})(R_{q}-R_{4})}{R_{q}R_{2}})V_{C}-\frac{V_{c}}{R_{1}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{s^{2}C^{2}R_{f1}R_{f2}R_{q}R_{1}+sCR_{1}R_{2}R_{f2}(R_{1}+R_{2})(R_{q}-R_{4})+R_{q}R_{2}}{R_{q}R_{1}R_{2}}V_{C}
$$

$$
\frac{V_{c}}{V_{in}} = -\frac{1}{R_{g}}\frac{R_{q}R_{1}R_{2}}{s^{2}C^{2}R_{f1}R_{f2}R_{q}R_{1}+CR_{1}R_{2}R_{f2}(R_{1}+R_{2})(R_{q}-R_{4})s+R_{q}R_{2}}
$$

$$
\frac{V_{c}}{V_{in}} = -\frac{R_{q}R_{1}R_{2}}{C^{2}R_{g}R_{f1}R_{f2}R_{q}R_{1}}\frac{1}{s^{2}+\frac{CR_{1}R_{2}R_{f2}(R_{1}+R_{2})(R_{q}-R_{4})}{C^{2}R_{f1}R_{f2}R_{q}R_{1}}s+\frac{R_{q}R_{2}}{C^{2}R_{f1}R_{f2}R_{q}R_{1}}}
$$

$$
\frac{V_{c}}{V_{in}} = -\frac{R_{1}}{R_{g}}\frac{\frac{R_{2}}{C^{2}R_{f1}R_{f2}R_{1}}}{s^{2}+\frac{R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{CR_{f1}R_{q}}s+\frac{R_{2}}{C^{2}R_{f1}R_{f2}R_{1}}}
$$

Parametros:
$$
K = \frac{R_{1}}{R_{g}}~~~~~~~~\frac{\omega_{0}}{Q}=\frac{R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{CR_{f1}R_{q}}~~~~~~~~\omega_{0}^{2}=\frac{R_{2}}{C^{2}R_{f1}R_{f2}R_{1}}
$$

## High Pass Filter

$$
\frac{V_{in}}{R_{g}} = \frac{V_{x}}{R_{1}}-\frac{V_{c}}{R_{1}}+\frac{V_{x}}{R_{2}}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = \frac{R_{1}(R_{q}-R_{4})}{R_{q}}V_{b}+\frac{1}{SCR_{f2}R_{1}}V_{b}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}}V_{b}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = (\frac{R_{1}(R_{q}-R_{4})}{R_{q}}+\frac{1}{SCR_{f2}R_{1}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}})V_{b}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = -(\frac{R_{1}(R_{q}-R_{4})}{R_{q}}+\frac{1}{SCR_{f2}R_{1}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}})\frac{1}{sCR_{f1}}V_{a}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = -(\frac{R_{1}(R_{q}-R_{4})}{R_{q}}+\frac{1}{SCR_{f2}R_{1}}+\frac{R_{2}(R_{q}-R_{4})}{R_{q}})\frac{1}{sCR_{f1}}V_{a}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{SCR_{f2}R_{1}(R_{1}+R_{2})(R_{q}-R_{4})+R_{q}}{SCR_{q}R_{f2}R_{1}}\frac{1}{sCR_{f1}}V_{a}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{SCR_{f2}R_{1}(R_{1}+R_{2})(R_{q}-R_{4})+R_{q}}{S^{2}C^{2}R_{q}R_{f1}R_{f2}R_{1}}V_{a}-\frac{V_{a}}{R_{2}}
$$

$$
\frac{V_{in}}{R_{g}} = -\frac{S^{2}C^{2}R_{q}R_{f1}R_{f2}R_{1}+SCR_{f2}R_{1}R_{2}(R_{1}+R_{2})(R_{q}-R_{4})+R_{q}R_{2}}{S^{2}C^{2}R_{q}R_{f1}R_{f2}R_{1}}V_{a}
$$

$$
\frac{V_{a}}{V_{in}} = -\frac{1}{R_{g}}\frac{S^{2}C^{2}R_{q}R_{f1}R_{f2}R_{1}}{S^{2}C^{2}R_{q}R_{f1}R_{f2}R_{1}+SCR_{f2}R_{1}R_{2}(R_{1}+R_{2})(R_{q}-R_{4})+R_{q}R_{2}}
$$

$$
\frac{V_{a}}{V_{in}} = -\frac{1}{R_{g}}\frac{S^{2}}{S^{2}+\frac{CR_{f2}R_{1}R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{C^{2}R_{q}R_{f1}R_{f2}R_{1}}s+\frac{R_{q}R_{2}}{C^{2}R_{q}R_{f1}R_{f2}R_{1}}}
$$

$$
\frac{V_{a}}{V_{in}} = -\frac{1}{R_{g}}\frac{S^{2}}{S^{2}+\frac{R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{CR_{f1}R_{q}}s+\frac{R_{2}}{C^{2}R_{f1}R_{f2}R_{1}}}
$$

Parametros:
$$
K = \frac{1}{R_{g}}~~~~~~~~\frac{\omega_{0}}{Q}=\frac{R_{2}(R_{1}+R_{2})(R_{q}-R_{4})}{CR_{f1}R_{q}}~~~~~~~~\omega_{0}^{2}=\frac{R_{2}}{C^{2}R_{f1}R_{f2}R_{1}}
$$
