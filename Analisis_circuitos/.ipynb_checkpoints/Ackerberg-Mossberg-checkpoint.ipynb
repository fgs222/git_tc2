# Ackerberg-Mossberg

Ecuaciones:

$$\frac {V_{1}}{\frac{R}{k}}+\frac{V_{2}}{R_{1}} = - \frac{SQRC+1}{QR}V_{a}~~~~~~(1)$$


$$\frac{V_{2}}{R}=-\frac{V_{b}}{R}~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(2)$$   


$$\frac{V_{a}}{R_{1}}=-\frac{V_{b}}{\frac{1}{SC}}~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(3)$$   


De (2):

$$V_{b} = -V_{2}$$

De (3):

$$V_{a} = -SR_{1}C V_{b}$$

## Low Pass Filter

Remplazando (2) y (3) en (1):

$$
\frac{k}{R}V_{1}=-\frac{SQRC+1}{QR}(-SR_{1}C)(-V_{2})- \frac{V_{2}}{R_{1}}
$$

$$
-\frac{SQRC+1}{QR}(-SR_{1}C)(-V_{2})- \frac{V_{o}}{R_{1}}=\frac{k}{R}V_{1}
$$

$$
\frac{-s^{2}QRR_{1}C^{2}-SR_{1}C}{QR}V_{2}-\frac{V_{2}}{R_{1}} = \frac{k}{R}V_{1}
$$

$$
\frac{-s^{2}QRR_{1}^{2}C^{2}-SR_{1}^{2}C-QR}{QRR_{1}}V_{2}= \frac{k}{R}V_{1}
$$

$$
\frac{V_{2}}{V_{1}} = \frac{-k}{R}\frac{QRR_{1}}{-s^{2}QRR_{1}^{2}C^{2}-SR_{1}^{2}C-QR}
$$

$$
\frac{V_{2}}{V_{1}} = \frac{-k}{R}\frac{QRR_{1}}{QRR_{1}^{2}C^{2}}\frac{1}{s^{2}+\frac{SR_{1}^{2}C}{QRR_{1}^{2}C^{2}}+\frac{QR}{QRR_{1}^{2}C^{2}}}
$$

$$
\frac{V_{2}}{V_{1}} = \frac{-kR_{1}}{R}\frac{\frac{1}{R_{1}^{2}C^{2}}}{s^{2}+\frac{1}{QRC}s+\frac{1}{R_{1}^{2}C^{2}}}
$$

Parametros:

$$\omega_{0}^{2} = \frac{1}{R_{1}^{2}C^{2}}~~~~~~~~~~~~~~~~~~~~~~\frac{\omega_{0}}{Q} = \frac{1}{QRC}~~~~~~~~~~~~~~~~~~~~~~K = \frac{kR_{1}}{R}$$

## Band Pass Filter

Remplazando (2) y (3) en (1):

$$
\frac {V_{1}}{\frac{R}{k}} = - \frac{SQRC+1}{QR}V_{a}-\frac{1}{SR_{1}^{2}C}V_{a}
$$

$$
\frac{-S^{2}QRR_{1}^{2}C^{2}-SR_{1}^{2}C-QR}{SQRR_{1}^{2}C}V_{a}=\frac{k}{R}V_{1} 
$$

$$
\frac{V_{a}}{V_{1}}=\frac{-k}{R}\frac{SQRR_{1}^{2}C}{S^{2}QRR_{1}^{2}C^{2}+SR_{1}^{2}C+QR}
$$

$$
\frac{V_{a}}{V_{1}}=\frac{-k}{R}\frac{QRR_{1}^{2}C}{QRR_{1}^{2}C^{2}}\frac{S}{S^{2}+\frac{R_{1}^{2}C}{QRR_{1}^{2}C^{2}}s+\frac{QR}{QRR_{1}^{2}C^{2}}}
$$

$$
\frac{V_{a}}{V_{1}}=\frac{-k}{R}\frac{S}{S^{2}+\frac{1}{QRC}s+\frac{1}{R_{1}^{2}C^{2}}}
$$

Parametros:

$$\omega_{0}^{2} = \frac{1}{R_{1}^{2}C^{2}}~~~~~~~~~~~~~~~~~~~~~~\frac{\omega_{0}}{Q} = \frac{1}{QRC}~~~~~~~~~~~~~~~~~~~~~~K = \frac{k}{R}$$
